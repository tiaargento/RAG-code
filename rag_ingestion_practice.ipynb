RAG Ingestion Practice Using MarkitDown

```python
from pathlib import Path
from markitdown import MarkItDown

DATA_DIR = Path(r"C:\rag_ingestion_practice\data")

md = MarkItDown()

# First test: convert a plain text sample
file_path = DATA_DIR / "sample.txt.txt"
print("Testing sample file:", file_path)
try:
	result = md.convert(file_path)
	print("SAMPLE_CONVERT_OK", "markdown_len=", len(getattr(result, 'markdown', '')))
	print(result.markdown[:1000])
except Exception:
	import traceback
	traceback.print_exc()

# If sample produced no output, try OCR on the PDF
pdf_path = DATA_DIR / "MEDRPT-2024-PAT-3847_medical_report_scan.pdf"
if not (hasattr(result, 'markdown') and len(result.markdown) > 0):
	print("Sample produced empty output, trying OCR on PDF:", pdf_path)
	try:
		result = md.convert(pdf_path, ocr=True)
		print("OCR_CONVERT_OK", "markdown_len=", len(getattr(result, 'markdown', '')))
		print(result.markdown[:1000])
	except Exception:
		import traceback
		traceback.print_exc()
```
